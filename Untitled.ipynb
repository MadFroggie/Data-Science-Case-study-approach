# Boolean Masks for data cleaning 
To help clean the data, we would need logical mask also called as Boolean mask. A logical mask is a way to filter an array, or a series, by some condition. We can find the values that clear a certain condition and the output of such comparison will be Boolean values. We'll be trying this out on synthetic data first

```python
from numpy.random import default_rng
```

Now we use a seed for the random number generator, if we set the seed to some number then we'll get the same results from the random numer generator everytime we run it

```python
rg = default_rng(12345)
```

```python
random_integers = rg.integers(low=1,high=5, size=100)
```

```python
random_number = random_integers[random_integers!=3]
print("Before", random_integers)
print("After", random_number)
```

Continuing verification of data integrity: we will examine some of the duplicate IDs we discovered. We will now locate these IDs and examine them. Then we take action to remove rows of dubious quality from the dataset

```python
duplicates = id_counts==2
duplicates
```

```python
duplicate_indexes = id_counts.index[duplicates]
duplicate_indexes
```

Using the first three IDs on our list of dupes, dupe_ids[0:3], we will plan to
first find the rows containing these IDs. If we pass this list of IDs to the .isin
method of the ID Series, this will create another logical mask we can use on the
larger DataFrame to display the rows that have these IDs. The .isin method
is nested in a .loc statement indexing the DataFrame in order to select the
location of all rows containing True in the Boolean mask. The second argument
of the .loc indexing statement is :, which implies that all columns will be
selected. By performing the following steps, we are essentially filtering the
DataFrame in order to view all the columns for the first three duplicate IDs.

```python
df.loc[df['ID'].isin(duplicate_indexes[0:5]),:]
```

Now you can see that there's one column with the correct data and another with 0 this could be because of a faulty join condition. We remove the columns that have 0 in the data

```python
df_zero_mask=df==0
```

```python
features_wo_zero = df_zero_mask.iloc[:,1:].all(axis=1)
sum(features_wo_zero)
```

The above output tells us that 315 entries have zeros for every column but the first one. 

```python
df_clean = df.loc[~features_wo_zero,:].copy()
```

# One hot encoding (OHE) 
This is a way of encoding categorical variables , which may consist of text labels in the raw data, into a numercal feature that can be used in mathematical models. 

## Example of implementing OHE for a categorical feature

```python
df_clean_2['EDUCATION_CAT'] = 'none' #Creating a new column called education and assigning none to all values
df_clean_2['EDUCATION_CAT']
```

```python
cat_mapping = {1: "graduate school", 2: "university", 3: "high school", 4: "others"}
```

```python
df_clean_2['EDUCATION_CAT'] = df_clean_2['EDUCATION'].map(cat_mapping)
df_clean_2['EDUCATION_CAT']
```

```python
edu_ohe = pd.get_dummies(df_clean_2['EDUCATION_CAT'])
edu_ohe.head(10)
```

you can now see why this is called the one hot encoding: across all these columns there's only 1 in only 1 column and the rest are 0's. Now we will concatenate this new dataframe with the original one and examine the results. We will be using the concat function to concatenate the two dataframes the axis will be =1, which will concatenate them horizontally, along the column axis.

```python
df_with_ohe = pd.concat([df_clean_2, edu_ohe], axis=1)
df_with_ohe[['EDUCATION_CAT', 'graduate school','high school', 'university', 'others']].head(10)
```

OHE is another way to encode
categorical features that avoids the implied numerical structure of an ordinal
encoding. However, notice what has happened here: we have taken a single
column, EDUCATION, and exploded it out into as many columns as there were
levels in the feature. In this case, since there are only four levels, this is not such
a big deal. However, if your categorical variable had a very large number of
levels, you may want to consider an alternate strategy, such as grouping some
levels together into single categories.
